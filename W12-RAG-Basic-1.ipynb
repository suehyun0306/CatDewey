# 🔍 RAG에 대한 이해

## 1. RAG란 무엇인가?

RAG는 **Retrieval-Augmented Generation**의 약자로, 말 그대로
**검색(Retrieval)**과 **생성(Generation)**을 결합한 AI 활용 방식입니다.

일반적인 LLM(ChatGPT 등)은 모델이 학습한 시점까지의 지식만을 이용해 답변을 생성합니다.
이 경우 최신 정보나 실제 문서에 있는 구체적인 규정, 숫자, 통계 등을 제대로 반영하지 못하거나,
존재하지 않는 내용을 그럴듯하게 만들어 내는 **환각(hallucination)** 문제가 발생할 수 있습니다.

RAG는 이런 한계를 보완하기 위해, **먼저 외부 문서에서 관련 내용을 검색**한 뒤,
그 검색 결과를 LLM에 전달하여 **문서 기반으로 답변을 생성**하게 만드는 구조입니다.

한 줄로 요약하면:

> LLM에게 “그냥 아는 척하지 말고, **먼저 문서를 찾아보고** 답해라”라고 시키는 방식입니다.

## 2. LLM의 한계와 RAG의 필요성

### 2.1 LLM의 장점
- 자연어 이해와 생성 능력이 뛰어나다.
- 글쓰기, 요약, 번역, 아이디어 브레인스토밍, 코드 작성 등 다양한 작업에 활용 가능하다.

### 2.2 LLM의 한계: 환각(Hallucination)
- LLM은 **사실 여부를 검증하는 구조**가 아니라, **문맥상 그럴듯한 다음 단어를 예측**하는 구조이다.
- 이 때문에 실제로 존재하지 않는 정보도 매우 설득력 있게 만들어낼 수 있다.
- 예시:
  - 존재하지 않는 학과를 만들어 설명하거나,
  - 학교 규정을 임의로 지어내어 설명하는 경우 등.

### 2.3 RAG의 필요성
- 우리는 “말 잘하는 AI”뿐만 아니라 **“정확한 AI”**를 필요로 한다.
- 특히 행정, 학사, 의료, 법률 등에서는 **근거가 있는 답변**이 중요하다.
- RAG는 LLM이 **실제 문서(규정집, 매뉴얼, 공지, 보고서 등)**를 참고하도록 만들어,
  환각 문제를 줄이고 최신 정보를 반영할 수 있게 해주는 기술이다.

## 3. RAG의 기본 구조

RAG의 기본적인 동작 흐름은 다음과 같습니다.

1. **질문 입력**: 사용자가 자연어로 질문을 한다.
2. **검색(Retrieval)**: 질문과 의미적으로 관련 있는 문서 조각(Chunk)을 검색한다.
3. **생성(Generation)**: 검색해온 문서들을 LLM에 함께 전달하고, 그 내용을 바탕으로 답변을 생성한다.

이를 그림으로 표현하면:

**질문 → (임베딩) → 관련 문서 검색 → 문서 + 질문을 LLM에 입력 → 답변 생성**

일반 LLM과의 차이는, **답변을 생성하기 전에 반드시 검색 과정을 거친다는 점**입니다.

## 4. 검색, 임베딩, Chunking

### 4.1 키워드 기반 검색
- 네이버, 구글, 도서관 검색처럼 **단어 일치**를 기반으로 문서를 찾는다.
- 표현이 조금만 달라져도 검색이 어려울 수 있다. (예: “전기차” vs “EV”)

### 4.2 의미 기반 검색(벡터 검색)
- 문장이나 문서를 **임베딩(embedding)**이라는 숫자 벡터로 변환한다.
- 의미가 비슷한 문장일수록 벡터 공간에서 서로 가깝게 위치한다.
- 질문도 임베딩으로 바꾸고, 벡터 공간에서 가장 가까운 문서 임베딩을 찾는다.

### 4.3 Chunking
- 실제 문서는 매우 길 수 있다 (여러 페이지짜리 PDF 등).
- 문서를 적당한 크기의 **조각(Chunk)**으로 나누어 각각 임베딩한다.
- Chunk 크기는 너무 작아도, 너무 커도 안 좋고, **내용과 모델 특성에 맞는 적절한 크기**를 찾는 것이 중요하다.

## 5. RAG vs 파인튜닝, 그리고 활용 사례

### 5.1 파인튜닝(Fine-Tuning)
- 모델 자체를 다시 학습시켜 새로운 지식을 내부에 주입하는 방식이다.
- 많은 데이터, 시간, 비용이 필요하다.
- 한 번 학습하면 지식을 바꾸기 어렵다.

### 5.2 RAG
- 모델은 그대로 두고, **참고할 문서만 교체**한다.
- 문서만 업데이트해도 도메인 지식이 즉시 반영된다.
- 비용과 유지보수 측면에서 훨씬 유리하다.

### 5.3 대학과 공공에서의 활용 예시
- 학사 규정 Q&A 챗봇 (휴학, 복학, 장학금, 수강신청 등).
- 행정·법령 안내 챗봇 (주소, 민원, 공공서비스 안내 등).
- 전공 강의자료 기반 학습 도우미.
- 공공데이터 또는 기업 내부 문서를 기반으로 한 업무 지원 챗봇.

## 6. 이 노트북에서 할 실습

이제 아래 코드 셀부터는 **아주 작은 학사 규정 데이터**를 가지고
간단한 RAG 시스템을 구현해 보는 실습을 진행합니다.

구체적으로는:

1. 몇 개의 학사 관련 문장을 `documents` 리스트로 정의합니다.
2. 각 문장을 임베딩하여 벡터로 변환합니다.
3. 사용자의 질문을 임베딩하고, 코사인 유사도 기반으로 가장 관련 있는 문서를 찾습니다.
4. 검색된 문서를 LLM에 함께 전달하여, **문서 기반 답변**을 생성합니다.

이 과정을 통해, RAG가 **검색 + 생성**의 조합이라는 것을 직접 확인할 수 있습니다.

## 9. 예시 문서 정의 (학사 규정 텍스트)

```python
documents = [
    {
        'id': 1,
        'title': '휴학 규정',
        'text': '''우리 대학교의 일반 휴학은 최대 3년까지 가능하다.
정상 학기 중에는 학사일정에 따라 정해진 기간에만 휴학 신청이 가능하며,
군휴학은 별도 규정을 따른다.'''
    },
    {
        'id': 2,
        'title': '복학 규정',
        'text': '''복학은 휴학 기간이 종료되는 학기의 개강 전 지정 기간에 신청해야 한다.
군휴학자의 경우 전역일을 기준으로 복학 신청 시기를 정하며,
복학 신청을 하지 않을 경우 제적 등의 불이익이 발생할 수 있다.'''
    },
    {
        'id': 3,
        'title': '장학금 안내',
        'text': '''성적 장학금은 직전 학기 평점이 3.5 이상인 학생을 대상으로 심사를 진행하며,
소득분위, 봉사활동, 지도교수 추천 등을 종합적으로 고려한다.
장학금 신청은 학교 포털을 통해 학기마다 진행된다.'''
    },
    {
        'id': 4,
        'title': '수강신청 안내',
        'text': '''수강신청은 매 학기 정해진 기간에 온라인으로 진행되며,
학년별로 신청 가능 시간대가 다를 수 있다.
수강정정 기간에는 신청 과목을 변경할 수 있으나,
최소·최대 학점 기준을 반드시 지켜야 한다.'''
    },
]

len(documents)
```

## 10. 임베딩 및 코사인 유사도 함수 정의

```python
import numpy as np

def get_embedding(text: str):
    """주어진 텍스트를 OpenAI 임베딩 벡터로 변환합니다."""
    response = client.embeddings.create(
        model='text-embedding-3-small',
        input=text
    )
    return response.data[0].embedding

def cosine_similarity(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# 모든 문서에 대해 임베딩 미리 계산
for doc in documents:
    doc['embedding'] = get_embedding(doc['text'])

print('문서 개수:', len(documents))
print('첫 번째 문서 임베딩 길이:', len(documents[0]['embedding']))
```

## 11. Retrieval 함수: 질문과 가장 비슷한 문서 찾기

```python
def retrieve_relevant_documents(query: str, k: int = 2):
    """질문을 임베딩하고, 코사인 유사도 기준 상위 k개 문서를 반환합니다."""
    query_embedding = get_embedding(query)

    scored_docs = []
    for doc in documents:
        score = cosine_similarity(query_embedding, doc['embedding'])
        scored_docs.append((score, doc))

    scored_docs.sort(key=lambda x: x[0], reverse=True)
    top_docs = [doc for score, doc in scored_docs[:k]]
    return top_docs

# 테스트
test_query = '휴학은 최대 몇 년까지 할 수 있어?'
top_docs = retrieve_relevant_documents(test_query, k=2)
for i, d in enumerate(top_docs, 1):
    print(f'[{i}]', d['title'])
```

## 12. RAG 답변 생성 함수

```python
def answer_with_rag(question: str, k: int = 2) -> str:
    """RAG 방식으로 질문에 답변합니다."""
    # 1. 관련 문서 검색
    top_docs = retrieve_relevant_documents(question, k=k)

    # 2. 컨텍스트 텍스트 구성
    context_blocks = []
    for doc in top_docs:
        block = f"[{doc['title']}]\n{doc['text']}"
        context_blocks.append(block)
    context_text = '\n\n---\n\n'.join(context_blocks)

    # 3. 사용자 프롬프트 구성
    user_prompt = f'''
당신은 '우리 대학교 학사 도우미 챗봇'입니다.
아래에 주어진 '참고 문서'의 내용만을 바탕으로, 사용자 질문에 한국어로 친절하게 답변하세요.
- 문서에 없는 내용은 추측하지 말고 모른다고 답변하세요.
- 가능한 경우, 규정의 핵심 숫자나 조건을 함께 알려주세요.

[참고 문서]
{context_text}

[사용자 질문]
{question}
'''.strip()

    # 4. LLM 호출
    response = client.chat.completions.create(
        model='gpt-4.1-mini',  # 계정 상황에 맞게 조정 가능
        messages=[
            {"role": "system", "content": "당신은 대학 행정 정보를 도와주는 조교입니다."},
            {"role": "user", "content": user_prompt},
        ],
        temperature=0.2,
    )

    answer = response.choices[0].message.content
    return answer

# 예시 질문
print(answer_with_rag('일반 휴학은 최대 몇 년까지 가능한가요?', k=2))
```

## 13. 여러 질문에 대해 RAG 동작 확인하기

```python
questions = [
    '일반 휴학은 최대 몇 년까지 가능한가요?',
    '군 휴학 후 복학 신청은 언제 해야 하나요?',
    '성적 장학금은 어떤 기준으로 받을 수 있나요?',
    '수강신청은 언제, 어떻게 해야 하나요?',
]

for q in questions:
    print('\n' + '='*60)
    print('질문:', q)
    print('-'*60)
    print(answer_with_rag(q, k=2))
```

## 14. (선택) RAG 없이 일반 LLM 답변과 비교

```python
def answer_without_rag(question: str) -> str:
    response = client.chat.completions.create(
        model='gpt-4.1-mini',
        messages=[
            {"role": "system", "content": "당신은 일반적인 질문에 답하는 어시스턴트입니다."},
            {"role": "user", "content": question},
        ],
        temperature=0.7,
    )
    return response.choices[0].message.content

q = '우리 학교에서 일반 휴학은 최대 몇 년까지 가능한가요?'
print('질문:', q)
print('\n[일반 LLM 답변]')
print(answer_without_rag(q))
print('\n[RAG 기반 답변]')
print(answer_with_rag(q, k=2))
```
